```python
import json
import pandas as pd
from collections import Counter

# Load the JSON data
file_path = r"C:\Users\spatt\Desktop\namingnames\docanalysis\output\graph_data_20250809_060817.json"

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Convert to DataFrame for easier analysis
df = pd.DataFrame(data)

df.head()
```

```python
# Filter for 1946 data
df_1946 = df[df['year'] == 1946]

print(f"Total records in dataset: {len(df)}")
print(f"Records for 1946: {len(df_1946)}")
print()

# Count frequency of target countries in 1946
target_counts = df_1946['target'].value_counts()

print("Most frequent targets in 1946:")
print("=" * 40)
for target, count in target_counts.items():
    print(f"{target}: {count}")

print()
print("Top 10 most targeted countries in 1946:")
print(target_counts.head(10))
```

```python
# Filter for all records targeting Germany (DEU) in 1946
germany_targets = df_1946[df_1946['target'] == 'DEU']

print(f"Total critiques of Germany in 1946: {len(germany_targets)}")
print("=" * 50)

# Display each source-target relationship with Germany
for idx, row in germany_targets.iterrows():
    print(f"\nSource: {row['source']}")
    print(f"Rhetorical Move: {row['rhetorical_move']}")
    print(f"Full Text: {row['full_text']}")
    print(f"Tags: {row['tags']}")
    print("-" * 50)

# Quick analysis of unique sources and rhetorical patterns
print("\n" + "=" * 50)
print("SUMMARY ANALYSIS:")
print("=" * 50)

sources = germany_targets['source'].tolist()
print(f"Sources targeting Germany: {sources}")

print("\nUnique rhetorical themes:")
for idx, move in enumerate(germany_targets['rhetorical_move'], 1):
    print(f"{idx}. {move}")

# Extract all tags for pattern analysis
all_tags = []
for tags_list in germany_targets['tags']:
    all_tags.extend(tags_list)

tag_counts = Counter(all_tags)
print(f"\nMost common tags across all Germany critiques:")
for tag, count in tag_counts.most_common():
    print(f"  {tag}: {count}")
```

# Yearly target analysis 

```python
def analyze_year_targets(df, year: int, top_x: int = 5):
    """
    Analyze the top X most targeted countries for a given year
    
    Args:
        df: DataFrame with diplomatic data
        year: Year to analyze
        top_x: Number of top targets to analyze in detail
    """
    print(f"DIPLOMATIC TARGETS ANALYSIS FOR {year}")
    print("=" * 60)
    
    # Filter for specified year
    df_year = df[df['year'] == year]
    
    if len(df_year) == 0:
        print(f"No records found for year {year}")
        return
    
    print(f"Total records in dataset: {len(df)}")
    print(f"Records for {year}: {len(df_year)}")
    print()
    
    # Count frequency of target countries
    target_counts = df_year['target'].value_counts()
    
    print(f"All targets in {year}:")
    print("-" * 40)
    for target, count in target_counts.items():
        print(f"{target}: {count}")
    
    print()
    print(f"TOP {top_x} DETAILED ANALYSIS:")
    print("=" * 60)
    
    # Get top X targets
    top_targets = target_counts.head(top_x).index.tolist()
    
    for i, target in enumerate(top_targets, 1):
        print(f"\n{i}. TARGET: {target} ({target_counts[target]} critiques)")
        print("=" * 50)
        
        # Filter for this target
        target_data = df_year[df_year['target'] == target]
        
        # Display each critique
        for idx, row in target_data.iterrows():
            print(f"\nSource: {row['source']}")
            print(f"Rhetorical Move: {row['rhetorical_move']}")
            print(f"Full Text: {row['full_text']}")
            print(f"Tags: {row['tags']}")
            print("-" * 30)
        
        # Summary analysis for this target
        sources = target_data['source'].tolist()
        print(f"\nSources targeting {target}: {sources}")
        
        print(f"\nRhetorical themes for {target}:")
        for j, move in enumerate(target_data['rhetorical_move'], 1):
            print(f"  {j}. {move}")
        
        # Tag analysis
        all_tags = []
        for tags_list in target_data['tags']:
            all_tags.extend(tags_list)
        
        tag_counts_target = Counter(all_tags)
        print(f"\nMost common tags for {target}:")
        for tag, count in tag_counts_target.most_common():
            print(f"  {tag}: {count}")
        
        if i < len(top_targets):  # Don't print separator after last target
            print("\n" + "=" * 60)

# Example usage:
# analyze_year_targets(df, 1946, 5)  # Top 5 targets for 1946
# analyze_year_targets(df, 1947, 3)  # Top 3 targets for 1947
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt

# Path to your JSONL file
file_path = r"C:\Users\spatt\Desktop\namingnames\classdeductive\run_01\results.jsonl"

# Read JSONL and parse "output" field into year and classification
rows = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        rec = json.loads(line)
        output_str = rec.get("output", "")

        # Remove <analysis> tags and parse JSON inside
        output_str = output_str.replace("<analysis>", "").replace("</analysis>", "").strip()
        try:
            output_json = json.loads(output_str)
            rows.append({
                "year": output_json.get("year"),
                "classification": output_json.get("classification")
            })
        except json.JSONDecodeError:
            pass

df = pd.DataFrame(rows).dropna(subset=["year", "classification"])

# Create plot
plt.figure(figsize=(12, 6))
df.groupby(["year", "classification"]).size().unstack(fill_value=0).plot(
    kind="bar", stacked=False, figsize=(12, 6)
)
plt.title("Distribution of Classifications by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title="Classification")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt

# Path to your JSONL file
file_path = r"C:\Users\spatt\Desktop\namingnames\classdeductive\run_01\results.jsonl"

# Read JSONL and parse "output" field into year and classification
rows = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        rec = json.loads(line)
        output_str = rec.get("output", "")

        # Remove <analysis> tags and parse JSON inside
        output_str = output_str.replace("<analysis>", "").replace("</analysis>", "").strip()
        try:
            output_json = json.loads(output_str)
            rows.append({
                "year": output_json.get("year"),
                "classification": output_json.get("classification")
            })
        except json.JSONDecodeError:
            pass

df = pd.DataFrame(rows).dropna(subset=["year", "classification"])

# Group by year and classification, count, then normalize to proportions
proportions = (
    df.groupby(["year", "classification"])
    .size()
    .groupby(level=0)
    .apply(lambda x: x / x.sum())  # normalize within each year
    .unstack(fill_value=0)
)

# Plot as stacked proportions
proportions.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Proportion of Classifications by Year")
plt.xlabel("Year")
plt.ylabel("Proportion")
plt.legend(title="Classification", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Classification Analysis 

```python
import pandas as pd
import json

file_path = r"C:\Users\spatt\Desktop\namingnames\classdeductive\run_01\results.jsonl"

# 1) Load the JSONL as-is (keeps all original columns)
df = pd.read_json(file_path, lines=True)

# 2) Parse year and classification from the "output" field and add as new columns
def parse_output(output_str):
    if not isinstance(output_str, str):
        return pd.Series([None, None])
    cleaned = output_str.replace("<analysis>", "").replace("</analysis>", "").strip()
    try:
        parsed = json.loads(cleaned)
        return pd.Series([parsed.get("year"), parsed.get("classification")])
    except json.JSONDecodeError:
        return pd.Series([None, None])

df[["year", "classification"]] = df["output"].apply(parse_output)

# Optional: make year integer dtype (nullable) and tidy classification text
df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
df["classification"] = df["classification"].astype("string")

# 3) Now filter rows that have the required fields present
required = ["source", "target", "year", "classification", "item_id", "chunk_id", "doc_id"]
df = df.dropna(subset=required, how="any")

print(df.head())
print(df.dtypes)
```
